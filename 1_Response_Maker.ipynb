### Creates responses for Weekly Discussions

## Instructions

- Enter the course "session" in 'yyyy_mm' format, "week" in 'w' format, and run this cell!

```python
session = '2022_06'
week = '4'

#########################

import pandas as pd
import random as rand

students_df = pd.read_csv('..//..//..//Documents//0BSData/' +session+ '//Nicknames.csv', sep='\t')
df = pd.read_csv('..//..//..//Documents//0BSData//ResponseMaker' +week+ '.csv', sep='\t')

for nn in students_df.loc[:,'Nickname']:
    r = [nn]
    for i in range(len(df.iloc[0,:])):
        r.append(rand.choice(df.iloc[:,i]))
    print('Hello {},\n\n    {}  {}  {}  {}\n\n{}\n\nJonathan\n\n\n\n'.format(*r))
```
